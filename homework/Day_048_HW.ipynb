```python
import os
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# 設定 data_path
dir_data = './data/'

f_app = os.path.join(dir_data, 'train.csv')
train = pd.read_csv(f_app,header=None)

b_app = os.path.join(dir_data, 'trainLabels.csv')
train_label = pd.read_csv(b_app,header=None)

c_app = os.path.join(dir_data, 'test.csv')
test = pd.read_csv(c_app,header=None)

# 建立模型
clf = GradientBoostingClassifier()

# 訓練模型
clf.fit(train, train_label)

# 預測測試集
y_pred = clf.predict(test)
```

```python
data = {'Id':range(1, 9001),
       'Solution':y_pred}
df = pd.DataFrame(data)
df.to_csv("Day_048_HW.csv",index=False)
```
